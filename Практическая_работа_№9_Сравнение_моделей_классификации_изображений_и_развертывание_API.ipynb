```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import os
import cv2

# Загрузка моделей
models = {
    "2lab": load_model(drive_path + '2lab/2lab_car_mt_bus.keras'),
    "3lab": load_model(drive_path + '3lab/car_mt_bus.keras'),
    "4lab": load_model(drive_path + '4lab/car_bus_mt_bn.keras'),
    "5lab": load_model(drive_path + '5lab/tm_car_bus_mt.keras')
}

# Размеры входов для каждой модели
input_sizes = {
    "2lab": (64, 64),  # и в вектор
    "3lab": (128, 128),
    "4lab": (128, 128),
    "5lab": (64, 64)
}

# Классы (при условии, что порядок одинаков во всех моделях)
classes = {0: 'car', 1: 'motorbike', 2: 'bus'}

def predict_with_all_models(img_path, models):
    """Предсказывает класс для изображения используя все модели"""
    try:
        # Загрузка изображения и нормализация
        img = image.load_img(img_path)
        img_array = image.img_to_array(img) / 255.0

        results = []
        for model_name, model in models.items():
            try:
                # Получаем нужный размер для данной модели
                size = input_sizes.get(model_name, (128, 128))
                img_resized = cv2.resize(img_array, size)

                # Подготовка изображения
                if model_name == "2lab":
                    # Преобразуем в вектор, как ожидает модель
                    img_processed = img_resized.reshape(1, -1)
                else:
                    # Оставляем как 4D тензор: (1, H, W, 3)
                    img_processed = np.expand_dims(img_resized, axis=0)

                # Предсказание
                pred = model.predict(img_processed, verbose=0)
                class_idx = np.argmax(pred[0])
                confidence = np.max(pred[0])
                results.append(f"{model_name}: {classes[class_idx]} ({confidence:.2%})")

            except Exception as e:
                print(f"[Ошибка] Модель '{model_name}': {str(e)}")
                continue

        # Вывод результатов
        print(f"\nРезультаты классификации для: {os.path.basename(img_path)}")
        for res in results:
            print(f"- {res}")

        # Отображение изображения
        plt.figure(figsize=(5, 5))
        plt.imshow(img)
        plt.axis('off')
        plt.title("Входное изображение")
        plt.show()

        return results

    except Exception as e:
        print(f"[Ошибка обработки изображения] {img_path}: {str(e)}")
        return None

# Путь к тестовым изображениям
image_path = os.path.join(drive_path, "3lab", "15_test")

# Список всех изображений
image_extensions = ('.jpg', '.jpeg', '.JPG', '.JPEG')
image_files = [os.path.join(image_path, f) for f in os.listdir(image_path)
               if f.lower().endswith(image_extensions) and os.path.isfile(os.path.join(image_path, f))]

# Предсказания по всем изображениям
for img_path in sorted(image_files):
    predict_with_all_models(img_path, models)
```

```python
import numpy as np
import pandas as pd
import time
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
import seaborn as sns

# Хранилище для метрик
results_metrics = []

# Хранилище предсказаний для построения confusion matrix
all_predictions = {model_name: [] for model_name in models}
all_true = []

# Предсказания по всем изображениям
for img_path in sorted(image_files):
    filename = os.path.basename(img_path)
    y_true_label = None
    if "car" in filename:
      y_true_label = "car"
    if "moto" in filename:
      y_true_label = "motorbike"
    if "bus" in filename:
      y_true_label = "bus"
    if y_true_label is None:
        print(f"Пропущен {filename} — нет метки")
        continue
    all_true.append(y_true_label)

    img = image.load_img(img_path)
    img_array = image.img_to_array(img) / 255.0

    for model_name, model in models.items():
        try:
            size = input_sizes.get(model_name, (128, 128))
            img_resized = cv2.resize(img_array, size)

            if model_name == "2lab":
                img_processed = img_resized.reshape(1, -1)
            else:
                img_processed = np.expand_dims(img_resized, axis=0)

            start_time = time.time()
            pred = model.predict(img_processed, verbose=0)
            infer_time = time.time() - start_time

            class_idx = np.argmax(pred[0])
            pred_label = classes[class_idx]
            confidence = np.max(pred[0])

            all_predictions[model_name].append(pred_label)

            results_metrics.append({
                'model': model_name,
                'filename': filename,
                'true': y_true_label,
                'pred': pred_label,
                'confidence': confidence,
                'inference_time': infer_time
            })

        except Exception as e:
            print(f"[Ошибка] {model_name} на {filename}: {e}")

# Преобразуем результаты в DataFrame
df = pd.DataFrame(results_metrics)

# Подсчет метрик по каждой модели
summary = []
for model_name in models.keys():
    preds = df[df['model'] == model_name]
    y_true = preds['true']
    y_pred = preds['pred']

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
    rec = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
    avg_inf_time = preds['inference_time'].mean()

    summary.append({
        'Model': model_name,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-score': f1,
        'Avg Inference Time (s)': avg_inf_time
    })

df_summary = pd.DataFrame(summary)
print("\nСводная таблица метрик:\n", df_summary)

# Визуализация метрик
df_summary.set_index('Model')[['Accuracy', 'Precision', 'Recall', 'F1-score']].plot(kind='bar', figsize=(10, 6), ylim=(0, 1), title='Сравнение моделей по метрикам')
plt.ylabel("Значение")
plt.grid(True)
plt.tight_layout()
plt.show()

# Визуализация времени
df_summary.set_index('Model')[['Avg Inference Time (s)']].plot(kind='bar', figsize=(8, 4), color='orange', title='Среднее время инференса')
plt.ylabel("Секунды")
plt.grid(True)
plt.tight_layout()
plt.show()

# Матрицы ошибок
for model_name in models.keys():
    y_true = df[df['model'] == model_name]['true']
    y_pred = df[df['model'] == model_name]['pred']
    cm = confusion_matrix(y_true, y_pred, labels=list(classes.values()))
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=classes.values(), yticklabels=classes.values(), cmap="Blues")
    plt.title(f'Матрица ошибок для модели: {model_name}')
    plt.xlabel("Предсказано")
    plt.ylabel("Истинно")
    plt.tight_layout()
    plt.show()

# Лучшая модель по F1
best_model = df_summary.sort_values(by='F1-score', ascending=False).iloc[0]
print(f"\nЛучшая модель по F1-мере: {best_model['Model']} (F1 = {best_model['F1-score']:.2%})")
```
